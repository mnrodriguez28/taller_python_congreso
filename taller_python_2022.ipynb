**Indicadores**

image.png

La mediana, que divide a la caja en dos, facilitar la comprensin de si la distribucin es simtrica o asimtrica.

Si la mediana se sita en el centro de la caja --> la distribucin es **simtrica** y tanto la media, mediana y moda coinciden.

Si la mediana corta la caja en dos lados desiguales se tiene:

- **Asimetra positiva** o segada a la derecha si la parte ms larga de la caja es la parte superior a la mediana. Los datos se concentran en la parte inferior de la distribucin. La media suele ser mayor que la mediana.
- **Asimetra negativa** o sesgada a la izquierda si la parte ms larga es la inferior a la mediana. Los datos se concentran en la parte superior de la distribucin. La media suele ser menor que la mediana.

La Escala de Estado de Flow tiene *36 items*.\
A su vez, esos items se dividen en *9 subescalas* (4 items por subescala).\
De este modo, ahora tenemos que crear nuevas columnas que refieran a los puntajes promedio de cada una de las subescalas de estado de flow. 

Primero vamos a establecer cules son los indices ("nmeros") de columna que refieren a las preguntas de cada subescala. Para hacer esto vamos a crear listas de indices. Recuerden que las listas se crean con corchetes "[]".

````python
flow1_index = [2, 11, 20, 29] # Balance entre nivel de habilidad y desafio
flow2_index = [3, 12, 21, 30] # Fusin entre la accin y la consciencia
flow3_index = [4, 13, 22, 31] # Claridad de los objetivos
flow4_index = [5, 14, 23, 32] # Concentracin en la tarea
flow5_index = [6, 15, 24, 33] # Claridad del feedback
flow6_index = [7, 16, 25, 34] # Sensacin de control
flow7_index = [8, 17, 26, 35] # Prdida de la autoconciencia
flow8_index = [9, 18, 27, 36] # Distorsin del paso del tiempo
flow9_index = [10, 19, 28, 37] # Experiencia autotlica
````

Bien! Ahora creamos una a una las columna con los valores promedio de cada subescala de Estado de Flow.

 **DISCLAIMER**  Podriamos haber implementado un "for loop" para crear estas variables de forma mas eficiente. 

````python
data['flow_1'] = data.iloc[:,flow1_index].mean(axis =1)
data['flow_2'] = data.iloc[:,flow2_index].mean(axis =1)
data['flow_3'] = data.iloc[:,flow3_index].mean(axis =1)
data['flow_4'] = data.iloc[:,flow4_index].mean(axis =1)
data['flow_5'] = data.iloc[:,flow5_index].mean(axis =1)
data['flow_6'] = data.iloc[:,flow6_index].mean(axis =1)
data['flow_7'] = data.iloc[:,flow7_index].mean(axis =1)
data['flow_8'] = data.iloc[:,flow8_index].mean(axis =1)
data['flow_9'] = data.iloc[:,flow9_index].mean(axis =1)
````

Nos fijamos si se crearon estas nuevas variables:

````python
data.head(3)
````

Ahora, agregamos una columna '**flow_total**' que contenga el promedio de todos las 9 subescalas de estado de flow (de cada participante).

````python
data['flow_total'] = data.loc[:,'flow_1':'flow_9'].mean(axis =1)
````

Nuevamente, nos fijamos si se cre esa nueva variable `flow_total`:

````python
data.head(3)
````

Definimos ahora el puntaje de corte a partir del cual se considera que el participante entr en Estado de Flow:

````python
puntaje_de_corte = 6.8
````

Ya establecida la variable `puntaje_de_corte`, creamos una nueva columna con el promedio total de flow de cada sujeto.

Utilizamos ahora un mtodo de *NumPy* para aplicar un condicional al otorgarle el valor a una nueva columna:

`np.where(condicion, valor si la condicion es True, valor si la condicion es False)`

````python
data['supera_puntaje_de_corte'] = np.where(data['flow_total']>puntaje_de_corte, 'Si', 'No')
````

Nos fijamos si se creo la variable `supera_puntaje_de_corte`:

````python
data.head(3)
````

#### **Estadistica inferencial:** 

**Objetivo**  Testear si el rendimiento de los participantes que entraron en Estado de Flow es mejor en comparacin con quienes no entran en Estado de Flow.

Primero vamos a evaluar esta comparacin con grficos!

````python
plt.figure(figsize = (7,5))
sns.boxplot(y = 'K/D Ultima Partida', x = 'supera_puntaje_de_corte', data = data)
plt.title('Rendimiento segn Estado de Flow', fontweight = 'bold')
plt.ylabel('Rendimiento (K/D)')
plt.xlabel('Estado de Flow')
````

Efectivamente a simple vista parece haber una diferencia entre grupos.\
Esta diferencia es estadsticamente significativa?

Realizamos una prueba **T de Student** para muestras independientes para responder a este interrogante.

*La prueba T de dos muestras (tambin llamada prueba t de muestras independientes) es un mtodo utilizado para probar si las medias de poblacin desconocidas de dos grupos son iguales o no.*

*Guardamos los resultados de este test en las variables:*
- '*summary*': describe cada uno de los grupos
- '*results*': devuelve el resultado del anlisis estadstico 

````python
import researchpy as rp
summary, results = rp.ttest(group1 = data['K/D Ultima Partida'][data['supera_puntaje_de_corte'] == 'Si'], group1_name = "Flow", 
                            group2 = data['K/D Ultima Partida'][data['supera_puntaje_de_corte'] == 'No'], group2_name = "No Flow")
````

````python
print(summary)
````

````python
print(results)
````

Cmo reportamos esto?! 


**Interpretacin**   La proporcin K/D (es decir, el rendimiento) para jugadores de CS GO que entraron en Estado en Flow, M=1.72 (1.40, 2.03), fue estadsticamentemayor que en jugadores de CS GO que no entraron en estado de flow , M= 0.69 (0.47, 0.90); *t*(36)= 3.35, *p*= 0.0001.

 Todavia tenemos que **chequear los supuestos** del test estadstico empleado!

1.   **Distribuciones normales**

````python
import scipy.stats as stats
stats.shapiro(data['K/D Ultima Partida'][data['supera_puntaje_de_corte'] == 'Si'])
````

````python
stats.shapiro(data['K/D Ultima Partida'][data['supera_puntaje_de_corte'] == 'No'])
````

Las distribuciones parecen ser nomales!

2.   **Homogeneidad (u homocedasticidad) de las varianzas**  

Otro supuesto es que las varianzas entre grupos sean iguales.

La varianza es una medida de dispersin que representa la variabilidad de una serie de datos respecto a su media.
De esta forma, se utiliza el test de Levene para testear dicho supuesto:

````python
stats.levene(data['K/D Ultima Partida'][data['supera_puntaje_de_corte'] == 'Si'],
             data['K/D Ultima Partida'][data['supera_puntaje_de_corte'] == 'No'],
             center = 'mean')
````

Como el resultado de la prueba es *significativo*, esto indicara que **los grupos tienen varianza distinta** y que **la prueba T puede no ser el mejor mtodo estadstico a utilizar**. En este caso, sera mejor utilizar el **Test de Welch**!
